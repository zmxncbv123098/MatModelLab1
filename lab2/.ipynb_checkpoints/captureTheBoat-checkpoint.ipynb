```python
import matplotlib.pyplot as plt
import math 
import numpy as np
from scipy.integrate import odeint
from matplotlib.pyplot import polar
print("Setup complete!")
```

```python
k = 18.4 # Начальное расстояние 
v = 4.6
fi = 3*math.pi/4
```

```python
# Движение охраны 
def dr(r, tetha):
    res = r/math.sqrt((v**2)-1)
    return res

# Движение браконьеров
def f2(t):
    xt = math.tan(fi)*t
    return xt
```

```python
"""
Получается, что х1 и х2 в моем случае будут равняться k/(4.6+-1)
'+-' - это значит плюс минус, так как у меня два случая.

"""

for vx in [3.6, 5.6]:
    r0 = k/vx # Наши два случая 
    
    tetha = np.arange(0, 2*math.pi, 0.01)
    
    r = odeint(dr, r0, tetha) # Решение дифф. уравнения
    t = np.arange(0, 15, 1)
    
    # Переведем всё в полярные координаты 
    pol = t*t + f2(t)*f2(t)
    r2 = np.sqrt(pol)
    
    tetha2 = (np.tan(f2(t)/t))**(-1)
    
    # Построим графики
    polar(tetha, r, "g")
    polar(tetha2, r2, "b")
```
