```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
a = 0.27 # коэффициент естественной смертности хищников
b = 0.37 # коэффициент естественного прироста жертв
c = 0.068 # коэффициент увеличения числа хищников
d = 0.064 # коэффициент смертности жертв

X = [9, 29] # начальное значение x и у (популяция хищников и популяция жертв)
```

```python
def funk(x, t):
    dx0 = -a*x[0] + c*x[0]*x[1]
    dx1 = b*x[1] - d*x[0]*x[1]
    return dx0, dx1
```

```python
t = np.arange(0, 100, 0.1)
y = odeint(funk, X, t)

y2 = y[:,1] # массив хищников 
y1 = y[:,0] # массив жертв
```

```python
plt.plot(t,y1, label='хищники')
plt.plot(t,y2, label='жертвы')
plt.legend()
```

```python
plt.plot(y1,y2) #построение графика зависимости изменения численности хищников от изменения численности жертв
plt.plot(12,36, 'ro', label='начальное состояние')
plt.plot(b/d,a/c, 'go', label='стационарное состояние')
plt.legend()
plt.grid(axis='both')
```
